```python
import requests
from io import BytesIO
from pypdf import PdfReader
import fitz
```

```python
url="https://op.europa.eu/o/opportal-service/download-handler?identifier=1d83c480-f353-4079-a9d1-b4b71fd64410&format=pdfa1b&language=en"
```

```python
#Old pdf reader, replace it with the new one

def pdf_reader_PdfReader(url):
    reader = PdfReader(BytesIO(requests.get(url).content))
    number_of_pages = len(reader.pages)
    pages = {}
    for page_number in range(number_of_pages):
        page = reader.pages[page_number]
        pages[f'page_{page_number + 1}'] = page.extract_text().strip()
    return pages
        
```

```python
pages=pdf_reader_PdfReader(url)
print(pages["page_1"])
```

```python
# new reader

def pdf_reader_fitz(url):
    request = requests.get(url)
    filestream = BytesIO(request.content)
    pdf = fitz.open(stream=filestream, filetype="pdf")
    pages={}
    for page in pdf:
        pages[f'page_{page.number+1}'] = page.get_text().strip()
    return pages
```

```python
pages=pdf_reader_fitz(url)
print(pages["page_1"])
```
